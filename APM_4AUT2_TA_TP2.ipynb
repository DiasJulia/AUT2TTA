Vehicle Suspension System

### **Equation of Motion:**

For a mass-spring-damper system, the equation of motion can be derived using Newton's second law or by applying the principles of dynamics. The general form of the equation is:

$$
m \frac{d^2x}{dt^2} + c \frac{dx}{dt} + kx = u + w
$$

### **State-Space Representation:**

To convert the second-order differential equation into a system of first-order differential equations, we introduce the following state variables:
- Let $ x_1 = x $ (position),
- Let $ x_2 = \dot x $ (velocity).

The system becomes:

$$\dot x = Ax+Bu+Dw$$

where
state vector $x = (position, velocity)$, and state and control matrices are the following:
$$ A = \begin{pmatrix} 0&1\\ -\frac{k}{m}&-\frac{c}{m}\end{pmatrix},\ B = \begin{pmatrix} 0\\ \frac{1}{m} 
\end{pmatrix},\ D =\begin{pmatrix} 0\\ \frac{1}{m} 
\end{pmatrix}$$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# x is a state vecot, t is a current time, A is a dynamic matrix, B is a control matrix, D is a disturbance matrix, 
# u_func(t) is a control input, w_func(t) is a disturbance input 
def StateSpace(x, t, A, B, D, u_func, w_func):
    w = w_func(t)
    u = u_func(t)
    return np.dot(A, x) + np.dot(B, u).flatten() + np.dot(D, w).flatten()

# OpenLoopInputOutputSystem with time-dependent w(t)
def OpenLoopInputOutputSystem(A, B, D, C, u_func, w_func, T, x0):
    # Solve the ODE system with odeint, passing the time-dependent w function
    solution = odeint(StateSpace, x0, T, args=(A, B, D, u_func, w_func))
    # Output calculation using C and the solution
    return (C @ solution.T) 

def BangBangController(A, B, D, C, u_const, ref, delay, w_func, T, x0):
    y_res = np.empty(x0.shape).reshape(-1,1)
    t0 = 0
    t1 = delay
    while t1 < T[-1]+delay:
        # Extract time segment
        T_bang = T[T<=t1]
        T = T[T_bang.shape[0]-1:]
        # Bang-Bang control logic
        u_func = lambda t: -u_const if C @ x0 - ref > 0 else (u_const if C @ x0 - ref < 0 else 0.0)
        # Compute system response
        sol = odeint(StateSpace, x0, T_bang, args=(A, B, D, u_func, w_func)).T
        x0 = sol[:,-1].flatten()
        # Update variables
        t0 = t1
        t1 += delay
        y_res = np.hstack((y_res, sol[:,1:]))
    return C @ y_res

def PIDController(A, B, D, C, Kp, Ki, Kd, ref, u_func, w_func, T, x0):
    y_res = np.empty(x0.shape).reshape(-1,1)
    integral = 0
    previous_error = 0
    for i in range(1, len(T)):
        t = T[i]
        dt = T[i] - T[i-1]
        # Calculate error
        error = ref - C @ x0
        integral += error * dt
        derivative = (error - previous_error) / dt if dt > 0 else 0
        # PID control input
        u_func = lambda t: Kp * error + Ki * integral + Kd * derivative
        # Compute system response for the current time step
        sol = odeint(StateSpace, x0, [T[i-1], T[i]], args=(A, B, D, u_func, w_func)).T
        x0 = sol[:,-1].flatten()
        previous_error = error
        y_res = np.hstack((y_res, sol[:,1:]))
    return C @ y_res
```

For this example, we will assume that the input of the system is the voltage source ($V$) applied to the motor's armature, while the output is the rotational speed of the shaft $\dot{\theta}$. We will assume that the magnetic field is constant and, therefore, that the motor torque is proportional (with constant $K_t$) to only the armature current. 2. Tune a PID controller using the Ziegler-Nichols Methods.

We have two ways of tunning, to use the first one, we must apply a step funcion to the open loop system, if the curve has an S-shape identity, we can calculate the constants based on the following proccess:

```python
def ziegler_nichols_s_shaped_response(T, y):
    func_aux = y[0,:]
    dfunc_aux = np.gradient(func_aux,T)
    d2func_aux = np.gradient(dfunc_aux,T)

    idx = (np.where(np.diff(np.sign(d2func_aux)) != 0)[0])

    if len(idx) == 0:
        print("No inflection points found.")
    if len(idx) > 0:
        idx = idx[0]  # Take the first inflection point

    print(f"Inflection points: {idx}")
    m = dfunc_aux[idx]

    n = func_aux[idx]-m*T[idx]
    L_nichols = -n/m

    x_T_nichols = (func_aux[-1]-n)/m
    T_nichols = x_T_nichols - L_nichols

    Kp = 1.2*T_nichols/L_nichols
    Ti = 2*L_nichols
    Td = 0.5*L_nichols
    Ki = Kp/Ti
    Kd = Kp*Td

    return Kp, Ki, Kd
```

## EX. 3 Inverted Pendulum

The system in this example consists of an inverted pendulum mounted to a motorized cart. The inverted pendulum system is an example commonly found in control system textbooks and research literature. Its popularity derives in part from the fact that it is unstable without control, that is, the pendulum will simply fall over if the cart isn't moved to balance it. Additionally, the dynamics of the system are nonlinear. The objective of the control system is to balance the inverted pendulum by applying a force to the cart that the pendulum is attached to. A real-world example that relates directly to this inverted pendulum system is the attitude control of a booster rocket at takeoff.



Let us consider the system with the following system parameters
    
    (M)       mass of the cart                         0.5 kg

    (b)       coefficient of friction for cart         0.1 N/m/sec

    (F)       force applied to the cart
    
    (y)       cart position coordinate
    
    (theta)   angle between the pendulum and the vertical axis

3) Design a PID controller corresponding to the reference signal $\theta_{ref}$ = 0 rad, while supposing that the angle is subject to a constant disturbance $w(t) = 0.1$

```python
def PIDControllerPendulum(A, B, D, C, Kp, Ki, Kd, ref, u_func, w_func, T, x0):
    y_res = np.empty(x0.shape).reshape(-1,1)
    integral = 0
    previous_error = 0
    for i in range(1, len(T)):
        t = T[i]
        dt = T[i] - T[i-1]
        # Calculate error
        error = ref - (C @ x0)[1]
        integral += error * dt
        derivative = (error - previous_error) / dt if dt > 0 else 0
        # PID control input
        u_func = lambda t: - (Kp * error + Ki * integral + Kd * derivative)
        # Compute system response for the current time step
        sol = odeint(StateSpace, x0, [T[i-1], T[i]], args=(A, B, D, u_func, w_func)).T
        x0 = sol[:,-1].flatten()
        previous_error = error
        y_res = np.hstack((y_res, sol[:,1:]))
    return C @ y_res
```

```python
parameters = {M: 0.5, m: 0.2, b: 0.1, l: 0.3, I: 0.006, g: 9.81}

A = np.array(jacobian_A.subs(parameters), dtype = float)
B = np.array(jacobian_B.subs(parameters), dtype = float)
C = np.array([
    [1, 0, 0, 0],
    [0, 0, 1, 0]
])
D = np.array([[0.0],[0.0],[0.0],[1.0]])

def w_func(t):
    return 0.1 # Disturbance input

ref = 0.0

x0 = np.array([0, 0, 0, 0]).flatten() # Initial state

t0 = 0  # Initial time
tf = 10  # Final time

T = np.linspace(t0, tf, 1000)  # Time vector
dt = T[1] - T[0]

kp = 100.0
ki = 40.0
kd = 20.0

y = PIDControllerPendulum(A, B, D, C, kp, ki, kd, ref, u_func, w_func, T, x0)

# Plot the response
plt.plot(T, y[1,:], linewidth=2.0, color='red')
plt.grid(color='black', linestyle='--', linewidth=1.0, alpha=0.7) 
plt.xlim([t0, tf]) 
plt.ylabel(r'Pendulum Angle (rad)')
plt.xlabel('Time (s)') 
plt.title('Inverted Pendulum Control')
plt.plot(T, np.ones(np.size(T))*ref, linewidth=2.0, color='blue') 
plt.legend(['PID controller response','reference input' ])
plt.show()
```

The simulation results show that while the PID controller successfully stabilizes the pendulum **angle** ($\theta$), the **cart position** ($x$) drifts indefinitely. This behavior is expected given the current control strategy and can be explained by the following factors:

* **SISO Control Limitation:** The implemented PID controller is a Single-Input Single-Output (SISO) system. It only monitors the error in the pendulum angle: $e(t) = \theta_{ref} - \theta(t)$. Because the position state is not part of the feedback loop, the controller does not attempt to bring the cart back to $x=0$.

* **System Coupling and Uncontrolled States:** In an inverted pendulum system, the position and angle are deeply coupled. By stabilizing the unstable angular pole without a secondary objective for the position, the cart's velocity results in a "pure integrator" effect.
